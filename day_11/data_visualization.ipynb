```python
#how to change the x and y axis titles?
g = sns.FacetGrid(kashti, hue='survived', col='sex', margin_titles=False,
     palette={1:'seagreen', 0:'grey'})
g =g.map(plt.scatter, 'fare', 'age', edgecolor='w').add_legend()
```

```python
g = sns.FacetGrid(kashti, hue='survived', col='sex', margin_titles=True,
     palette={1:'seagreen', 0:'grey'})
g =g.map(plt.scatter, 'fare', 'age', edgecolor='w').add_legend()
```

```python
g = sns.FacetGrid(kashti, hue='survived', col='who', margin_titles=True,
     palette={1:'seagreen', 0:'grey'})
g =g.map(plt.scatter, 'fare', 'age', edgecolor='w').add_legend()
```

```python
g = sns.FacetGrid(kashti, hue='survived', col='embark_town', margin_titles=False,
     palette={1:'seagreen', 0:'grey'})
g =g.map(plt.scatter, 'fare', 'age', edgecolor='w').add_legend()
```
