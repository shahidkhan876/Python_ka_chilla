```python
import pandas as pd
import seaborn as sns

#load the dataset
df = sns.load_dataset("titanic")

#fill the null values with the mean X=age
df["age"].fillna(df["age"].mean(), inplace=True)

X = df[["age"]]
y = df["fare"]

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV

#create a model
model = LinearRegression()

#define param_grid
param_grid = {"fit_intercept": [True, False]}

#object of GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring="neg_mean_squared_error")

grid_search.fit(X,y)
print("Best Parameters: ", grid_search.best_params_)
```

#KNN

```python
import pandas as pd
import seaborn as sns
import numpy as np

#load the dataset
df = sns.load_dataset("titanic")

#fill the null values with the mean X=age
df["age"].fillna(df["age"].mean(), inplace=True)

X = df[["age"]]
y = df["fare"]

from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

#create a model
model = KNeighborsRegressor()

#define param_grid
param_grid = {'n_neighbors':[1,3,5,7,9,11,13,15],'weights': ['uniform','distance']}

#object of GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring="precision")

grid_search.fit(X,y)
print("Best Parameters: ", grid_search.best_params_)
```

```python
import pandas as pd
import seaborn as sns
import numpy as np

#load the dataset
df = sns.load_dataset("titanic")

#fill the null values with the mean X=age
df["age"].fillna(df["age"].mean(), inplace=True)

X = df[["age"]]
y = df["fare"]

from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV

#create a model
model = KNeighborsRegressor()

#define param_grid
param_grid = {'n_neighbors':np.arange(1,30,2),'weights': ['uniform','distance']}

#object of GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring="neg_mean_squared_error")

grid_search.fit(X,y)
print("Best Parameters: ", grid_search.best_params_)
```

#DecisionTreeClassifier

```python
import pandas as pd
import seaborn as sns
import numpy as np

#load the dataset
df = sns.load_dataset("titanic")

#fill the null values with the mean X=age
df["age"].fillna(df["age"].mean(), inplace=True)

X = df[["age"]]
y = df["fare"]

from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import GridSearchCV

#create a model
model = DecisionTreeClassifier()

#define param_grid
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 10, ],
    'min_samples_split': [2, 5, 10]
}

#object of GridSearchCV
grid_search = GridSearchCV(model, param_grid, cv=5, scoring="neg_mean_squared_error")

grid_search.fit(X,y)
print("Best Parameters: ", grid_search.best_params_)
```
