```python
import pandas as pd
from ydata_profiling import ProfileReport  # or pandas_profiling if using an older version

# Create a sample DataFrame
data = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "Salary": [50000, 60000, 70000]
})

# Generate the profile report
profile = ProfileReport(data, title="Sample Data Report", explorative=True)

# To display the report in a Jupyter Notebook
profile.to_notebook_iframe()

# To save the report as an HTML file
profile.to_file("data_report.html")
```
