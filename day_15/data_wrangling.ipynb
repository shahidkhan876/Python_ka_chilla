# Step-3.1 Dealing with missing value
. remove the data

```python
df.drop(columns="deck", inplace = True)
```

```python
df.isnull().sum() / len(df) *100
```

```python
# mean fill karna ha
df['age'] = df.age.fillna(value = df['age'].mean)
```

```python
df.isnull().sum() / len(df) * 100
```

```python
df['embark_town'] = df.embark_town.fillna(value = df['embark_town'].mode)
```

```python
df.isnull().sum() / len(df) * 100
```

```python
df['embarked'] = df.embarked.fillna(value= df['embarked'].median)
```

```python
df.isnull().sum() / len(df) * 100
```

# 3.2.2 IQR

```python
Q1 = df['fare'].quantile(0.25)
Q3 = df['fare'].quantile(0.75)
IQR = Q3 - Q1
IQR

```

```python
#create a equation remove a outliers levels ky sath assignmnt
lower_bound = Q1 - 1.5 * IQR
upper_bound =Q3 + 1.5 * IQR
df = df[(df['fare']> lower_bound) & (df['fare']<upper_bound)]
df
```
